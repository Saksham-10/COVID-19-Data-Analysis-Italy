```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("covid-nationality.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
#extra columns dropped
data.drop("variation_total_confirmed",axis=1,inplace=True)
data.drop("casi_testati",axis=1,inplace=True)
```

```python
data.columns
```

```python
# check for the null values
data.isnull().sum()
```

```python
#we can see here that this is a linear realtionship as 
#people are recovering as the total cases are increasing 

sns.relplot(x='total_cases',y='recovered',data=data);
```

```python
#here we can se that the graph is exponential which means most of the cases
#are confirmed cases which is sad for italy
sns.relplot(x='total_cases',y='total_confirmed_cases',data=data); 
```

```python
#this is an interesting graph which explains us that the hospitalzied are  
#30K approx and confirmed cases are much higher than that so it dosent mean
#that if you dont have symptoms you dont have COVID-19

sns.relplot(x='total_confirmed_cases',y='hospitalized_with_symptoms',data=data);
```

```python
#here we use hue to differentiate between different catogrocial features

sns.relplot(x='total_confirmed_cases',y='hospitalized_with_symptoms',hue="recovered",data=data);
```

```python
sns.relplot(x='home_quarantine',y='total_cases',kind='line',data=data);
```

```python
#here we can say a positive side of home quarantine as we stay home the recovered also increases which is a linear relationship

sns.relplot(x='recovered',y='home_quarantine',kind='line',data=data);
```
